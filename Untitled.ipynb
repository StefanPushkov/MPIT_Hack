```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_excel('./data/train.xlsx')
```

```python
df.head()
```

```python
df1 = df[df['Дата']== pd.to_datetime('2016-01-01')]
```

```python
df4 = pd.read_excel('./data/train.xlsx', sheet_name= 'Координаты')
```

```python
df3 = pd.merge(df1, df4, how= 'inner', left_on= 'Скважина', right_on= '№ скважины')
```

```python
df3.columns
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x= df3['Координата X'], y= df3['Координата Y'], marker='o', c='r', edgecolor='b')

for i in df3.index:
    plt.annotate(df3['Характер работы'][i], xy=(df3['Координата X'][i], df3['Координата Y'][i]))

plt.show()
```
